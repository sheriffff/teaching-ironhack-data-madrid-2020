## A premier on algorithms



A coder is just someone that gives instructions to a computer. A set of instructions designed to solve an specific problem is called an "[Algorithm](https://en.wikipedia.org/wiki/Algorithm)" (yeap, nothing more, nothing less... just a very precise recipe to perform one task).

```python
# count sheep Kata
sheep = [True,  True,  True,  False,
          True,  True,  True,  True ,
          True,  False, True,  False,
          True,  False, False, True ,
          True,  True,  True,  True ,
          False, False, True,  None]

def count_sheeps(sheep):
    # solve this problem using a list comprehension!!
    return len([1 for i in sheep if i])

```

```python
count_sheeps(sheep)
```

What you did was to create an algorithm! You told the computer exactly what to do to accomplish a given task. How good is the algorithm in terms of scalability? Put it differently, how well will it perform if we feed it witha veeeeeeeeeeeery big list of sheep? We analyze that in terms of the "[Big O](https://en.wikipedia.org/wiki/Big_O_notation)" notation.

(**Note**: To illustrate the concepts we focus on the so-called "[time complexity](https://en.wikipedia.org/wiki/Time_complexity)". There is another type of complexity which is "[space complexity](https://en.wikipedia.org/wiki/Space_complexity)" that we do not cover here)

## Time complexity: "Big O" notation

**Lets try an experiment!**

First, your time to work a little ;)

```python
# knowing this...
random.choice([True, False])
```

```python
# complete the function so that it returns any number of sheeps
# (Yeap, you must use a list comprehension)
def create_sheeps(number:int):
    ''' create "number" of random True/False vales
        args:
            * number: integer representing the desired lenght
    '''
    return [random.choice([True, False]) for x in range(number)]
```

```python
create_sheeps(5)
```

```python
# lets create some sheeps! 
sheep_25 = create_sheeps(25)
sheep_50 = create_sheeps(50)
sheep_75 = create_sheeps(75)
sheep_100 = create_sheeps(100)
sheep_125 = create_sheeps(125)
sheep_150 = create_sheeps(150)
```

```python
%%timeit
count_sheeps(sheep_100)
```

```python
times = [1.69e-6, 2.76e-6, 4.73e-6, 4.84e-6, 6.96e-6, 7.56e-6]
number_of_sheep = [25, 50, 75, 100, 125, 150]
```

Lets see the relationship between the number of sheeps and the time it takes:

```python
plt.scatter(number_of_sheep, times);
plt.show()
```

Do you notices the linear pattern?? That is because the algorithm you wrote is O(n), that is: it scales linearly with the number of elements. 

Notice that even though the times of execution may vary from one computer to another. The number of operations that the computer needs to perform scale linearly with the size of the input.

### Bubble sort

A very naive and inefficient is called "[bubble sort](https://en.wikipedia.org/wiki/Bubble_sort#:~:text=Bubble%20sort%20has%20a%20worst,O(n%20log%20n)." (Still, it is a __valid__ solution)

```python
# Extracted from https://www.geeksforgeeks.org/bubble-sort/
def bubble_sort(arr):
    n = len(arr) 
  
    # Traverse through all array elements 
    for i in range(n): 
  
        # Last i elements are already in place 
        for j in range(0, n-i-1): 
  
            # traverse the array from 0 to n-i-1 
            # Swap if the element found is greater 
            # than the next element 
            
            
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 
                
    return arr
```

## Rules to remember

* What can cause time in a function?-
    * Operations (+, -, *, /)
    * Comparisons (<, >, ==)
    * Looping (for, while)
    * Function calls

* Rule 1: Think always of the worst possible case
* Rule 2: Constants usually dont count (unless the algorithm is constant itself!)
* Rule 3: Different inputs should have different variables. O(a+b). A and B arrays nested would be O(a*b)

    * \+ `for` sequential
    * \* `for` nested 
* Rule 4: Drop Non-dominant terms
